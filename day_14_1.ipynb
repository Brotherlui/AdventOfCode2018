```python
import numpy as np

def show(matrix, pos1, pos2, length):
    for i in range(length):
        if i == pos1:
            print(f'({matrix[i]})', end='')
        elif i == pos2:
            print(f'[{matrix[i]}]', end='')
        else:
            print(f' {matrix[i]} ', end='')
    print()


def rotate(pos, steps, length):
    return (pos + steps) % length


def scores(pos, matrix):
    print('Scores: ', end='')
    for i in range(pos, pos+10):
        print(f'{matrix[i]}', end='')
    print()


def run(num):
    recipes = np.zeros((30_000_000,), dtype=np.int)

    elf1 = 0
    elf2 = 1
    recipes[elf1] = 3
    recipes[elf2] = 7
    length = 2

    numstr = str(num)  
    part1 = part2 = False
    while not (part1 and part2):
        score = recipes[elf1] + recipes[elf2]
        einer = score % 10
        zehner = score // 10

        if zehner:
            recipes[length] = zehner
            length += 1
        recipes[length] = einer
        length += 1

        elf1 = rotate(elf1, 1+recipes[elf1], length)
        elf2 = rotate(elf2, 1+recipes[elf2], length)
        if not part1 and length >= num+10:
            scores(num, recipes)
            part1 = True

        last = ''.join(map(str, recipes[length-7:length]))
        if numstr in last:
            pos = last.index(numstr) + length - 7
            part2 = True

    print(f'Recipe string: {recipes[num:num+10]} at position {pos}')


run(598701)
```
