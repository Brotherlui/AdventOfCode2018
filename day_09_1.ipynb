```python
from time import time

class Marble():
    def __init__(self, value, clock=None, counter=None):
        self.current = True
        self.value = value
        self.clock = clock
        self.counter = counter

    def __str__(self):
        return f'Value:{self.value} Clock:{self.clock.value} Counter:{self.counter.value}'
    
    def show(self):
        if self.current:
            print(f'({self.value}) ', end='')
        else:
            print(f'{self.value} ', end='')
        clock = self.clock
        while clock.value != self.value:
            if clock.current:
                print(f'({clock.value}) ', end='')
            else:
                print(f'{clock.value} ', end='')
            clock = clock.clock
        print()
```

```python
max_player = 452
score = {key: 0 for key in range(1, max_player+1)}

circle = Marble(0)
circle.clock = circle
circle.counter = circle
current = circle

start_time = time()
player = 1
for marble in range(1, 7125000+1):
    if marble % 23 != 0:
        new = Marble(marble, clock=current.clock.clock, counter=current.clock)
        current.current = False
        current.clock.clock.counter = new
        current.clock.clock = new
        current = new
        # print(f'[{player}]  ', end='')
        # circle.show()
        player = player%max_player+1
    else:
        current.current = False
        # print(f'[{player}]  Marble {marble} ', end='')
        score[player] += marble
        
        remove = current
        for _ in range(7):
            remove = remove.counter
        score[player] += remove.value
        # print(f'{remove.value} will be removed and Player {player} has {score[player]} points')
        
        current = remove.clock
        current.current = True
        current.counter = remove.counter
        remove.counter.clock = current
        del remove
        
        # print(f'[{player}]  ', end='')
        # circle.show()
        player = player%max_player+1

print(f'Used time: {time()-start_time}')
best_player = max(score, key=lambda key: score[key])
print(f'Best player {best_player} with Score {score[best_player]}')
```
