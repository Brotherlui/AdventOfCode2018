```python
with open('day_18_input.txt', 'r') as datei:
    acres = [list(row.strip()) for row in datei]

width = len(acres[0]) + 2
height = len(acres) + 2

# create a matrix that has 1 extra cell to all sides
# and put the data into the center of it
grid = [[' ' for x in range(width)] for y in range(height)]
for y in range(1, height-1):
    for x in range(1, width-1):
        grid[y][x] = acres[y-1][x-1]


# a helper function to print the matrix on the screen
def show(acre):
    for y in range(1, len(acre)-1):
        for x in range(1, len(acre[0])-1):
            print(acre[y][x], end='')
        print()


# a helper function to make a deepcopy of the matrix
def copy(acre):
    return [[acre[y][x] for x in range(len(acre[0]))] for y in range(len(acre))]


# here we count all the neighbors for a single acre
def count_neighbors(acre, ypos, xpos):
    neighbors = {'.': 0, '|': 0, '#': 0, ' ': 0}
    for y in range(ypos-1, ypos+2):
        for x in range(xpos-1, xpos+2):
            neighbors[acre[y][x]] += 1
    neighbors[acre[ypos][xpos]] -= 1
    return neighbors['.'],  neighbors['|'], neighbors['#']


# a helper function to count the resources
def count_resources(grid):
    resources = {'#': 0, '|': 0, '.': 0}
    for y in range(1, height-1):
        for x in range(1, width-1):
            resources[grid[y][x]] += 1
    return resources['#'], resources['|']


# the main loop
def run(loop_count, grid):
    for loop in range(1, loop_count+1):
        magic = copy(grid)
        for y in range(1, height-1):
            for x in range(1, width-1):
                clear, trees, yards = count_neighbors(grid, y, x)
                current = grid[y][x]
                if current == '.' and trees >= 3:
                    magic[y][x] = '|'
                if current == '|' and yards >= 3:
                    magic[y][x] = '#'
                if current == '#' and (yards >= 1 and trees >= 1):
                    magic[y][x] = '#'
                if current == '#' and (yards < 1 or trees < 1):
                    magic[y][x] = '.'
        grid = copy(magic)

    show(grid)
    yards, trees = count_resources(grid)
    print(f'Total resources = {yards*trees}')
```

```python
run(10, grid)
```
