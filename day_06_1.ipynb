```python
import pandas as pd

def manhatten(a,b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])
```

```python
--- Part One ---

df = pd.read_csv('day_06_input.txt', names=['x', 'y'])

df.x -= df.x.min()
df.y -= df.y.min()

finite = []
for offset in [1, 5]:
    df.x += offset // 2
    df.y += offset // 2

    matrix = [[dict() for x in range(df.x.max()+1)] for y in range(df.y.max()+1)]
    coords = [(x,y,) for x in range(df.x.max()+1) for y in range(df.y.max()+1)]

    for p, point in enumerate(df.values):
        for coord in coords:
            dist = manhatten(point, coord)
            x, y = coord
            matrix[y][x][p] = dist

    count = dict()
    for x, y in coords:
        cell = matrix[y][x]
        minkey = min(cell, key=cell.get)
        minval = cell[minkey]
        delkeys = [key for key, val in cell.items() if val > minval]
        for key in delkeys:
            cell.pop(key, None)
        if len(cell) > 1:
            delkeys = [key for key, val in cell.items() if val == minval]
            for key in delkeys:
                cell.pop(key, None)
        else:
            for key, val in cell.items():
                try:
                    count[key] += 1
                except:
                    count[key] = 1
    finite.append(count)

delkeys = [key for key, val in finite[1].items() if finite[0][key] < val]
for key in delkeys:
    finite[1].pop(key, None)
print(finite[1][max(finite[1], key=lambda key: finite[1][key])]) # 6047
```

```python
# --- Part Two ---

df = pd.read_csv('day_06_input.txt', names=['x', 'y'])
df.x -= df.x.min()
df.y -= df.y.min()

coords = [(x,y,) for x in range(df.x.max()+1) for y in range(df.y.max()+1)]

distance = []
for coord in coords:
    dist_sum = 0
    for point in df.values:
        dist_sum += manhatten(coord, point)
    if dist_sum < 10000:
        distance.append(coord)

print(len(distance)) # 46320
```
